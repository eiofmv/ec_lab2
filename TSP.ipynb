# xqf131

```python
city_list = tsp.read_data('xqf131.tsp')
best_route = tsp.genetic_algorithm(population=city_list, population_size=150, 
                                   elite_size=50, mutation_rate=0.01, 
                                   generations=1500)

tsp.draw_route(best_route)
```
